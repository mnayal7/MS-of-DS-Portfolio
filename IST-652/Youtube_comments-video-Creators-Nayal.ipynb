```python
#making a list of how many likes a comment has if mentioning youtuber
coby_likes=(df.like_count[df.Coby_mentioned==1]).tolist()
cory_likes=(df.like_count[df.Cory_mentioned==1]).tolist()
garrett_likes=(df.like_count[df.Garrett_mentioned==1]).tolist()
cody_likes=(df.like_count[df.Cody_mentioned==1]).tolist()
tyler_likes=(df.like_count[df.Tyler_mentioned==1]).tolist()
none_mentioned_likes=((df.like_count[df.None_mentioned==1]).tolist())
```

```python
#making a dataframe with names mentioned and like in order to make boxplot
#starting with dictionary based on list of likes
d = dict(Coby= np.array(coby_likes),  Cory= np.array(cory_likes), Garrett= np.array(garrett_likes), Cody= np.array(cody_likes), Tyler= np.array(tyler_likes), No_one= np.array(none_mentioned_likes))
    
popularityDF=pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in d.items() ]))
popularityDF.Coby= popularityDF.Coby[popularityDF.Coby.notnull()].astype('int')
popularityDF.Cory= popularityDF.Cory[popularityDF.Cory.notnull()].astype('int')
popularityDF.Garrett= popularityDF.Garrett[popularityDF.Garrett.notnull()].astype('int')
popularityDF.Cody= popularityDF.Cody[popularityDF.Cody.notnull()].astype('int')
popularityDF.Tyler= popularityDF.Tyler[popularityDF.Tyler.notnull()].astype('int')
popularityDF.No_one=popularityDF.No_one.astype('int')
```

```python
#making one long dataframe of youtuber mentioned and number of likes of comment
#this step needed for anova
coby_likes_df=pd.DataFrame(list(zip(['Coby']*len(coby_likes), coby_likes)), columns =['Name', 'Likes'])
cory_likes_df=pd.DataFrame(list(zip(['Cory']*len(cory_likes), cory_likes)), columns =['Name', 'Likes'])
garrett_likes_df=pd.DataFrame(list(zip(['Garrett'], [0])), columns =['Name', 'Likes'])
cody_likes_df=pd.DataFrame(list(zip(['Cody']*len(cody_likes), cody_likes)), columns =['Name', 'Likes'])
tyler_likes_df=pd.DataFrame(list(zip(['Tyler']*len(tyler_likes), tyler_likes)), columns =['Name', 'Likes'])
none_mentioned_df=pd.DataFrame(list(zip(['None_Mentioned']*len(none_mentioned), none_mentioned)), columns =['Name', 'Likes'])
c1=coby_likes_df.append(cory_likes_df)
c2=c1.append(garrett_likes_df)
c3=c2.append(cody_likes_df)
c4=c3.append(tyler_likes_df)
long_likes_df=c4.append(none_mentioned_df)
print(type(long_likes_df))
long_likes_df.boxplot('Likes', by='Name', figsize=(12, 8))
```

```python
#creating one way anova of likes by name mentioned
import statsmodels.api as sm
from statsmodels.formula.api import ols
lm=ols('Likes ~ Name', data=long_likes_df).fit()
table=sm.stats.anova_lm(lm)
print(table)
```
