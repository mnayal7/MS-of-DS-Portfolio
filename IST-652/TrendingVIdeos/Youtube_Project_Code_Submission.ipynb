```python
#create rank column for ranking by latestest trending date as 1 by video id
def rank_date(df):
    df['Rank'] = df.groupby(by=["video_id"])["trending_date"].transform(lambda x: x.rank(ascending = False))
    return display(df.head())
```

```python
import string
import nltk
from nltk import FreqDist
from nltk.corpus import stopwords
```

```python
entertainmentDF=df_merge[df_merge.category_id=='Entertainment']
entertainmentDFUSA=df_merge[df_merge.Country=='United States']



#entertainment top Word List
clickBaitTitles=entertainmentDF.title.tolist()
#english top word string
clickBaitTitles=' '.join([i for i in clickBaitTitles if not i.isdigit()])
#get rid of numbers
translation = str.maketrans(string.ascii_letters, string.ascii_letters, string.digits)
clickBaitTitles=clickBaitTitles.translate(translation)

#remove punctuataion and make all lowercase
tokenizer = nltk.RegexpTokenizer(r"\w+")
clickBaitTokens = tokenizer.tokenize(clickBaitTitles)
clickBaitTokens = [t.lower( ) for t in clickBaitTokens]


#filtering out stop words
stopWords = set(stopwords.words('english'))
stopWordsArabic
stopWordsFrench
stopWordsGerman
moreArabicStopWords
clickBaitFiltered = []

for w in clickBaitTokens:
    if w not in stopWords and w not in stopWordsArabic and w not in stopWordsFrench and w not in stopWordsGerman and w not in moreArabicStopWords and w not in stopWordsSpanish and len(w)>2:
        clickBaitFiltered.append(w)

ClickBaitWordcount={}
for word in clickBaitFiltered:
  if word not in ClickBaitWordcount:
    ClickBaitWordcount[word] = 1
  else:
    ClickBaitWordcount[word] += 1

ClickBait_sorted_values = sorted(ClickBaitWordcount.values(), reverse=True) # Sort the values
ClickBait_wordcount_df=pd.DataFrame.from_dict(ClickBaitWordcount, orient='index')
ClickBait_wordcount_df=ClickBait_wordcount_df.sort_values(0, ascending=False).head(50)
ClickBait_wordcount_df=ClickBait_wordcount_df.reset_index()
ClickBait_wordcount_df.columns =['Entertainment_Words', 'Count']
ClickBait_wordcount_df.to_excel('entertainmentClickBait.xlsx')
display(ClickBait_wordcount_df.head(10))
```
